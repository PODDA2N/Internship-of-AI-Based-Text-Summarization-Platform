#### Create Main FastAPI App

```python
from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware
from routes.summarize import router

app = FastAPI()


app.add_middleware(                                  # Enable CORS (VERY IMPORTANT)
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)


app.include_router(router)                       # Include routes

@app.get("/")
def home():
    return {"message": "Backend is running"}
```

#### Why Create Main FastAPI App have you used and which purpose?

The backend of the AI-Based Text Summarization Platform is responsible for handling user requests, processing input data, and interacting with the AI model to generate summaries. It is developed using Python and FastAPI due to its high performance, simplicity, and built-in support for API development.

The main FastAPI application acts as the entry point of the backend system. It initializes the server, manages routing, and integrates different modules of the application.

FastAPI used for high performance (based on ASGI), automatic API documentation (/docs), easy integration with Python libraries, faster development compared to traditional frameworks also it connects all backend components and exposes APIs to the frontend.

#### Create API Routes

```python
from fastapi import APIRouter, UploadFile, File
from pydantic import BaseModel
from services.pdf_service import extract_text_from_pdf
from services.ai_service import generate_summary
from utils.text_cleaner import clean_text

router = APIRouter()


class TextRequest(BaseModel):                  # Request model
    text: str


@router.post("/summarize")                           #  Text Summarization
async def summarize_text(request: TextRequest):
    cleaned = clean_text(request.text)
    summary = generate_summary(cleaned)
    return {"summary": summary}


@router.post("/summarize-file")                              #  PDF Summarization
async def summarize_file(file: UploadFile = File(...)):
    text = extract_text_from_pdf(file)
    cleaned = clean_text(text)
    summary = generate_summary(cleaned)
    return {"summary": summary}
```

#### Why Create API Routes have you used and which purpose?

API routes define the endpoints through which the frontend communicates with the backend.

API Routes used for enables communication between frontend and backend, ensures structured and validated data transfer, supports modular and scalable design also routes act as bridges between user input and backend processing logic.

#### PDF Processing Service

```python
import pdfplumber
import io

def extract_text_from_pdf(file):
    text = ""

    with pdfplumber.open(io.BytesIO(file.file.read())) as pdf:
        for page in pdf.pages:
            if page.extract_text():
                text += page.extract_text()

    return text
```

#### Why PDF Processing Service have you used and which purpose?

This module extracts text content from uploaded PDF files so it can be processed by the AI model.

PDF Processing used for many real-world documents are in PDF format, AI models require plain text input, enables analysis of external documents.

PDF plumber used for accurate text extraction, supports multi-page PDFs, easy to integrate with Python also converts PDF content into readable text for further processing.

#### Text Cleaning

```python
def clean_text(text):
    return text.replace("\n", " ").strip()
```

#### Why Text Cleaning have you used and which purpose?

Text cleaning ensures that the input data is structured and noise-free before being sent to the AI model.

Text Cleaning used for improves input quality, helps AI generate accurate summaries, reduces unnecessary tokens and processing load also prepares raw input into a clean format suitable for AI processing.
